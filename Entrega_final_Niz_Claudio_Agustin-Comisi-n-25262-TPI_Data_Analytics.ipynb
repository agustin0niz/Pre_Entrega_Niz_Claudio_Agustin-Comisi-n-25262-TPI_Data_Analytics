```python
ventas_clean = df_ventas.copy()
clientes_clean = df_clientes.copy()
marketing_clean = df_marketing.copy()
```

**Punto 2.2**





```python
ventas_perf = (
    ventas_clean
    .assign(
        precio = ventas_clean["precio"].str.replace('$', '', regex=False).astype(float),
        ingreso = lambda x: x["precio"] * x["cantidad"]))

resumen_prod = (
    ventas_perf

    .groupby(
        by="producto",
        dropna=False,
        as_index=False,
        observed=False  )

    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('producto', 'count')))
print(resumen_prod.head(60))
```

**Punto 3.1**

```python
import matplotlib.pyplot as plt


df = resumen_prod.copy()
df_ordenado = df.sort_values("ingreso_total", ascending=False)

plt.figure(figsize=(12, 6))

colors = plt.cm.viridis(range(len(df_ordenado)))

# Gráfico de barras
barras = plt.bar(df_ordenado["producto"], df_ordenado["ingreso_total"], color=colors)

# Título y etiquetas
plt.title("Ingresos Totales por Producto", fontsize=16, weight="bold", color="darkblue")
plt.xlabel("Producto", fontsize=9)
plt.ylabel("Ingreso Total ($)", fontsize=9)

# Rotación de etiquetas
plt.xticks(rotation=45, ha="right")

for barra in barras:
    y = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        y / 2,
        f"${y:,.0f}",
        ha='center',
        va='center',
        fontsize=9,
        color="white",
        rotation=90
    )

plt.tight_layout()
plt.show()
```

```python
media = df['ingreso_total'].mean()
mediana = df['ingreso_total'].median()
moda = df['ingreso_total'].mode()

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
```

```python
ventas_cat = ventas_perf.copy()

resumen_cat = (
    ventas_cat
    .groupby(
        by="categoria",
        dropna=False,
        as_index=False
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        ventas=('ingreso', 'size'),
        precio_promedio=('precio', 'mean')
    )
    .sort_values(
        by='ingreso_total',
        ascending=False,
        na_position='last',
        ignore_index=True
    )
)

resumen_cat = resumen_cat.assign(
    ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas']
)

display(resumen_cat.head(20))
```

```python
import matplotlib.pyplot as plt

df = resumen_cat.copy()
df_ordenado = df.sort_values("ingreso_total", ascending=False)

plt.style.use("seaborn-v0_8-darkgrid")
plt.figure(figsize=(10, 6))

colors = plt.cm.plasma(range(len(df_ordenado)))

barras = plt.barh(df_ordenado["categoria"], df_ordenado["ingreso_total"], color=colors)

plt.title("Ingresos Totales por Categoría", fontsize=16, weight="bold", color="darkred")
plt.xlabel("Ingreso Total ($)", fontsize=12)
plt.ylabel("Categoría", fontsize=12)

for barra in barras:
    x = barra.get_width()
    plt.text(
        x + (x * 0.01),
        barra.get_y() + barra.get_height() / 2,
        f"${x:,.0f}",
        ha="left",
        va="center",
        fontsize=9,
        color="black"
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

ventas_perf['fecha_venta'] = pd.to_datetime(ventas_perf['fecha_venta'], errors='coerce')
ingresos_por_dia = ventas_perf.groupby('fecha_venta')['ingreso'].sum().sort_index()
ingresos_acumulados = ingresos_por_dia.cumsum()

plt.style.use("seaborn-v0_8-darkgrid")
plt.figure(figsize=(12, 6))

plt.plot(ingresos_por_dia, color='darkred', label='Ingreso diario', marker='o', markersize=4)
plt.plot(ingresos_acumulados, color='darkblue', label='Ingreso acumulado', linewidth=2, linestyle='--')

plt.title('Ingresos diarios y acumulados', fontsize=15, color="darkblue")
plt.xlabel('Fecha de venta')
plt.ylabel('Ingresos ($)')
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(ingresos_por_dia, color='darkorange', label='Ingreso diario', marker='o', markersize=4)
ax1.set_xlabel('Fecha de venta')
ax1.set_ylabel('Ingresos diarios ($)', color='darkorange')
ax1.ticklabel_format(style='plain', axis='y')

ax2 = ax1.twinx()
ax2.plot(ingresos_acumulados, color='darkred', label='Ingreso acumulado', linewidth=2, linestyle='--')
ax2.set_ylabel('Ingresos acumulados ($)', color='darkred')
ax2.ticklabel_format(style='plain', axis='y')

plt.title('Ingresos diarios vs acumulados', fontsize=15, color="darkred")
plt.grid(alpha=0.5, linestyle=':')
plt.tight_layout()
plt.show()
```

Esto se confirma con los valores de moda, donde hay varios ingresos repetidos entre $45K y $51K, y algunos mas extremos como $72K, $76K y $84K.  
La mediana nos da una idea mas realista del ingreso tipico por producto, mientras que la media refleja el impacto de los productos mas caros o exitosos.  

---

- Presencia de **outliers** y su impacto  

En los datos aparecen algunos ingresos muy por encima del resto. Estos outliers muestran que existen ventas puntuales con un ticket mucho mas alto que el promedio.  
- Que significan: son casos especiales que reflejan oportunidades de alto valor, pero no representan el comportamiento tipico de la mayoria de las ventas.  
- Impacto en el analisis: si se mezclan con el resto, pueden inflar la media y dar una vision demasiado optimista. La mediana, en cambio, se mantiene estable y refleja mejor el ingreso habitual.  
- Como aprovecharlos: conviene analizarlos aparte para entender que los genero (promos, fechas especiales, clientes premium) y ver si se pueden replicar esas condiciones.  

---

- Recomendaciones sobre **que productos potenciar** y **eficiencia de marketing**  

- Segmentacion por categoria:  
  - Electrodomesticos: destacar en campañas de practicidad y ahorro de tiempo.  
  - Electronica: apelar a innovacion y estilo de vida digital.  
  - Decoracion: aprovechar fechas especiales y tendencias de hogar.  
- Mediana como referencia: usarla para fijar objetivos realistas de ventas y precios, evitando que los outliers distorsionen la estrategia.  
- Outliers como oportunidad: analizar que disparo esas ventas excepcionales (promos, fechas clave, clientes premium) y replicar esas condiciones en campañas futuras.  
- Optimizacion temporal: reforzar marketing en dias de baja venta para suavizar la volatilidad diaria y aprovechar mejor el flujo constante que se ve en el acumulado.  
- Campañas cruzadas: combinar productos lideres con complementarios (ejemplo: microondas + accesorios de cocina, lampara + decoracion) para aumentar ticket promedio.
